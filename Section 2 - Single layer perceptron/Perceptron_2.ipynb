# Perceptron 2

## Inputs and weights

```
import numpy as np
```

```
inputs = np.array([35, 25])
```

```
type(inputs)
```

```
weights = np.array([0.8, 0.1])
```

## Sum function

```
# dot : 각 vector의 값을 곱하고, 더한다.
def sum(inputs, weights):
  return inputs.dot(weights)
```

```
s = sum(inputs, weights)
```

```
print(s)
```

## Step function

```
def step_function(sum):
  if sum >= 1:
    return 1
  return 0
```

## Final result

```
print(step_function(s))
```

```
weights[0] = -0.8
s = sum(inputs, weights)
print(step_function(s))
```
