weight(n+1) = weight(n) + (learning_rate(=fixed) * input * error)

## Inputs, outputs and weights

```
import numpy as np
```

```
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
```

```
inputs.shape
```

```
outputs = np.array([0, 0, 0, 1])
```

```
outputs.shape
```

```
weights = np.array([0.0, 0.0])
```

```
learning_rate = 0.1
```

## Step function

```
def step_function(sum):
  if sum >= 1:
    return 1
  return 0
```

## Calculate output

```
def calculate_output(instance):
  s = instance.dot(weights)
  return step_function(s)
```

```
print(calculate_output(np.array([[0,0]])))
```

```
print(calculate_output(np.array([[1,0]])))
```

```
print(calculate_output(np.array([[0,1]])))
```

```
print(calculate_output(np.array([[1,1]])))
```

## Train

```
def train():
  total_error = 1
  while total_error != 0:
    total_error = 0
    for i in range(len(outputs)):
      prediction = calculate_output(inputs[i])
      error = abs(outputs[i] - prediction)
      total_error += error
      if error > 0:
        for j in range(len(weights)):
          weights[j] += learning_rate * inputs[i][j] * error
          print('Weight updated:', weights[j])
    print('Total error: ', total_error)
```

```
train()
```

## Classification

```
print(calculate_output(np.array([0,0])))
print(calculate_output(np.array([1,0])))
print(calculate_output(np.array([0,1])))
print(calculate_output(np.array([1,1])))
```
